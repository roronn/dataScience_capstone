```python
import pandas as pd
```

```python
df = pd.read_csv('order_reviews_dataset.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df['review_comment_title'] = df['review_comment_title'].fillna('No comment title')
```

```python
df['review_comment_message'] = df['review_comment_message'].fillna('No comment message')
```

```python
df[['review_creation_date', 'review_answer_timestamp']] = df[['review_creation_date', 'review_answer_timestamp']].apply(pd.to_datetime)
```

```python
df.info()
```

```python
#check for duplicates

df.duplicated(subset=['review_id','order_id']).any()
```

```python
df.to_csv('order_reviews_dataset_clean.csv', index=True)
```
